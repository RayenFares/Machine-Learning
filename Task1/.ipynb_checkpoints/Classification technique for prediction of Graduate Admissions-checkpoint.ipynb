```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = dataset.drop(['Chance of Admit'], axis=1)
y = dataset['Chance of Admit']
#Standardization
scaler = StandardScaler()
X = scaler.fit_transform(X)
#X[['CGPA','GRE Score', 'TOEFL Score']] = scaler.fit_transform(X[['CGPA','GRE Score', 'TOEFL Score']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=48)
```

```python
lr=LinearRegression()
# Train the classifier on the training data
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
pd.DataFrame({"Actual": y_test, "Predict": y_pred}).head()
# Calculate accuracy
#accuracy = accuracy_score(y_test, y_pred)
#print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
lr_r2 = r2_score(y_test, y_pred)
lr_mse = mean_squared_error(y_test, y_pred)
lr_rmse = np.sqrt(lr_mse)
print('Linear Regression R2 Score: {0} \nLinear Regression MSE: {1}, \nLinear Regression RMSE:{2}'.format(lr_r2, lr_mse, lr_rmse))
```

Now if we want to make a classification model and see the accuracy we need to convert the chance of admission model so that 
if a candidate's Chance of Admit is greater than 80%, the candidate will receive the label 1.
If a candidate's Chance of Admit is less than or equal to 80%, the candidate will receive the label 0.

```python
y_train_1 = [1 if each > 0.8 else 0 for each in y_train]
y_test_1  = [1 if each > 0.8 else 0 for each in y_test]

# list to array
y_train_1 = np.array(y_train_1)
y_test_1 = np.array(y_test_1)
```

```python
print (y_train_1)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train,y_train_1)
```

```python
# we should normalize the data (the last error)
from sklearn.preprocessing import MinMaxScaler
scalerX = MinMaxScaler(feature_range=(0, 1))
X_train[X_train.columns] = scalerX.fit_transform(X_train[X_train.columns])
X_test[X_test.columns] = scalerX.transform(X_test[X_test.columns])
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train,y_train_1)
```

```python
y_predlogr = logr.predict(X_test)
print("Accuracy Score:", accuracy_score(y_predlogr, y_test_1))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_1,y_predlogr))
cm_lrc = confusion_matrix(y_test_1,y_predlogr)
```

```python
# cm visualization
f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm_lrc,annot = True,linewidths=0.5,linecolor="red",fmt = ".0f",ax=ax)
plt.title("Testing for Test Dataset")
plt.xlabel("predicted y values")
plt.ylabel("real y values")
plt.show()
```

Now let's test for train dataset

```python
cm_lrc_train = confusion_matrix(y_train_1,logr.predict(X_train))
f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm_lrc_train,annot = True,linewidths=0.5,linecolor="red",fmt = ".0f",ax=ax)
plt.xlabel("predicted y values")
plt.ylabel("real y values")
plt.title("Testing for Train Dataset")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score
print("precision_score: ", precision_score(y_test_1, y_predlogr))
print("recall_score: ", recall_score(y_test_1, y_predlogr))

from sklearn.metrics import f1_score
print("f1_score: ",f1_score(y_test_1, y_predlogr))
```
